## Graph MT to MS GDP

#### Firstly, open the path in where the file is located

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set pandas to not use scientific notation
pd.set_option('display.float_format', '{:.6f}'.format)
```

```python
# Load data from the specified Excel file and sheet
file_name = "States_GDP.xlsx"
sheet_name = 'Séries'
df = pd.read_excel(file_name, sheet_name=sheet_name)
df
```

```python
# Filter for Mato Grosso and Mato Grosso do Sul
states_to_plot = ['Mato Grosso', 'Mato Grosso do Sul']
df_filtered = df[df['Estado'].isin(states_to_plot)]

df_filtered
```

```python
# Get the years from 1980 to 2021 that are present in the dataframe column
#1980 is the first year available for IPCA to put in 2021 R$ values
years = [str(year) for year in range(1980, 2021)]
available_years = [year for year in years if year in df_filtered.columns]

# Reshape the data from wide to long format
df_long = pd.melt(df_filtered, id_vars=['Estado'], value_vars=available_years,
                    var_name='Ano', value_name='PIB')

# Convert 'Ano' to integer for correct sorting on the plot
df_long['Ano'] = df_long['Ano'].astype(int)

#Rename columns
df_long.rename(columns={'Ano': 'Year', 'Estado': 'State', 'PIB': 'GDP'}, inplace=True)

df_long
```

### Convert to 2021 values

```python
# Load and Prepare IPCA Data
# Assuming 'IPCA.xls' contains the data from your image
ipca_file = "IPCA.xls" 
ipca_df = pd.read_excel(ipca_file, sheet_name='Séries')

# Rename columns for clarity and consistency
ipca_df.rename(columns={'Data': 'Year', 'IPCA - taxa de variação - (% a.a.) - Instituto Brasileiro de Geografia e Estatística, Sistema Nacional de Índices de Preços ao Consumidor (IBGE/SNIPC) - PRECOS_IPCAG - ': 'IPCA'}, inplace=True)

# Convert IPCA percentage to a decimal factor for calculations (e.g., 5.91 -> 1.0591)
ipca_df['IPCA_factor'] = 1 + (ipca_df['IPCA'] / 100)
ipca_df
```

```python
# Calculate Cumulative Inflation Factor (Price Index)
# We sort by year to ensure the cumulative product is calculated correctly.
# The factor for 1980 is set to its own value to start the chain.
ipca_df.sort_values('Year', inplace=True)
ipca_df['Acumulated'] = ipca_df['IPCA_factor'].cumprod()
ipca_df
```

```python
# Merge GDP data with the inflation factor
# A left merge ensures we keep all GDP data and match the corresponding inflation factor for each year.
df_adj = pd.merge(df_long, ipca_df[['Year', 'Acumulated']], on='Year', how='left')

# Convert GDP and Accumulated columns to float
df_adj['GDP'] = pd.to_numeric(df_adj['GDP'], errors='coerce')
df_adj['Acumulated'] = pd.to_numeric(df_adj['Acumulated'], errors='coerce')

df_adj
```

```python
# Merge GDP data with the inflation factor
# A left merge ensures we keep all GDP data and match the corresponding inflation factor for each year.
df_adj = pd.merge(df_long, ipca_df[['Year', 'Acumulated']], on='Year', how='left')
df_adj
```

```python
# Get the 2021 inflation factor to use as the base
base_2021 = ipca_df[ipca_df['Year'] == 2021]['Acumulated'].iloc[0] if len(ipca_df[ipca_df['Year'] == 2021]) > 0 else ipca_df['Acumulated'].iloc[-1]

# Calculate GDP in 2021 R$ values
df_adj['GDP_2021'] = df_adj['GDP'] * (base_2021 / df_adj['Acumulated'])

df_adj
```

```python
# Create the plot
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Create the line plot with specific colors and line styles
for state in df_adj['State'].unique():
    state_data = df_adj[df_adj['State'] == state].sort_values('Year')
    
    if state == 'Mato Grosso':
        plt.plot(state_data['Year'], state_data['GDP_2021'] / 1000000, 
                 marker='o', linewidth=2.5, markersize=4, label=state, 
                 color='black', linestyle='-')
    elif state == 'Mato Grosso do Sul':
        plt.plot(state_data['Year'], state_data['GDP_2021'] / 1000000, 
                 marker='o', linewidth=2.5, markersize=4, label=state, 
                 color='gray', linestyle='--')

# Customize the plot
plt.xlabel('Year', fontsize=14)
plt.ylabel('2021 Million R$', fontsize=14)
plt.legend(fontsize=12, loc='lower center', bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.grid(True, alpha=0.3)

# Format the axes to remove scientific notation
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Remove scientific notation from y-axis
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Adjust layout to make room for legend below
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

# Save the plot with high resolution
plt.savefig('GDP_MT-MS.jpg', dpi=1200, bbox_inches='tight', format='jpg')

# Show the plot
plt.show()
```
